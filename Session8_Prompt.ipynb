```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace (0, 1, 100)
```

```python
def sin_func(x):
    return np.sin(x)

def cos_func(x):
    return np.cos(x)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, sin_func(x), color='lightblue')
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.subplot(1, 2, 2)
plt.plot(x, cos_func(x), color='lightpink')
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')

plt.savefig("Session8_Prompt.pdf")
plt.show()
```
