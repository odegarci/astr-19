```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_numbers = np.random.uniform(0, 1, 1000)

plt.figure(figsize=(12, 8))
plt.hist(random_numbers, bins=100, color='thistle', edgecolor='darkslateblue', alpha=0.5)

plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Histogram: Session 9')

plt.savefig("Session9_Histogram.pdf")
plt.show()
```
